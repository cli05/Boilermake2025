```python
import pandas as pd
```

```python
songList = pd.read_csv('dataset.csv')
print(len(songList))
```

```python
songListZero = songList[songList['duration_ms'] != 0] # gets rid of song with no duration
songListZero
```

```python
songListSort = songListZero.sort_values(by = 'track_id')
songListSorted = songListSort.reset_index(drop = True)
songListSorted.head(50)
```

```python
previousRow = None
for i, j in songListSorted.iterrows():
    currentRow = songListSorted.loc[i, 'track_id']
    if currentRow == previousRow:
        songListSorted.at[i, 'track_genre'] += "," + songListSorted.at[i - 1, 'track_genre']
        previousRow = currentRow
        songListSorted.drop(i - 1, inplace = True)
    else: 
        previousRow = currentRow
songListSorted.head(50)
```

```python
previousRow = None
songListCleaned = songListSorted
for i, j in songListSorted.iterrows():
    currentRow = songListSorted.loc[i, 'track_id']
    if currentRow == previousRow:
        songListCleaned.at[i - 1, 'track_genre'] += "," + songListCleaned.at[i, 'track_genre']
        previousRow = currentRow
        songListCleaned.drop(i, inplace = True)
songListCleaned.head(50)
```

```python
print(len(songListSorted))
```

```python
songListCleaned = songListSorted.reset_index(drop = True)
songListCleaned.head(10)
```
